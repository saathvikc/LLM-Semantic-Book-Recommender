```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
books['categories'].value_counts().reset_index().query("count > 50")
```

```python
category_mapping = {
    "Fiction": "Fiction",
    "Juvenile Fiction": "Children's Fiction",
    "Juvenile Nonfiction": "Children's Nonfiction",
    "Biography & Autobiography": "Nonfiction",
    "History": "Nonfiction",
    "Literary Criticism": "Nonfiction",
    "Science": "Nonfiction",
    "Self-Help": "Nonfiction",
    "Cooking": "Nonfiction",
    "Philosphy": "Nonfiction",
    "Religion": "Nonfiction",
    "Comics & Graphic Novels": "Fiction",
    "Drama": "Fiction",
    "Poetry": "Fiction",
}
```

```python
books["simple_categories"] = books["categories"].map(category_mapping)
books
```

```python
books[(books["simple_categories"].isna())]
```

```python
from transformers import pipeline
categories = [
    "Fiction",
    "Nonfiction"
]
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device="cuda")
```

```python
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
classifier(sequence, categories)
```

```python
import numpy as np

# Get the index of the highest scoring category
max_index = np.argmax(classifier(sequence, categories)["scores"])
max_label = classifier(sequence, categories)["labels"][max_index]
max_label
```

```python
def generate_simple_category(description, categories):
    result = classifier(description, categories)
    max_index = np.argmax(result["scores"])
    return result["labels"][max_index]
```

```python
from tqdm import tqdm

actual_categories = []
predicted_categories = []

for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
    predicted_categories += [generate_simple_category(sequence, categories)]
    actual_categories += ["Fiction"]    
```

```python
    
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
    predicted_categories += [generate_simple_category(sequence, categories)]
    actual_categories += ["Nonfiction"]
```

```python

predictions_df = pd.DataFrame({
    "actual": actual_categories,
    "predicted": predicted_categories
})

predictions_df
```

```python
predictions_df["correct"] = np.where(predictions_df["actual"] == predictions_df["predicted"], True, False)
predictions_df["correct"].value_counts()
predictions_df["correct"].mean()
```

```python
isbns = []
predictions = []

missing_categories = books.loc[books["simple_categories"].isna(), ["isbn13", "description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0, len(missing_categories))):
    sequence = missing_categories["description"][i]
    predictions += [generate_simple_category(sequence, categories)]
    isbns += [missing_categories["isbn13"][i]]
```

```python
missing_predictions_df = pd.DataFrame({"isbn13": isbns, "predictions": predictions})
missing_predictions_df
```

```python
books = pd.merge(books, missing_predictions_df, on="isbn13", how="left")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predictions"], books["simple_categories"])
books = books.drop(columns=["predictions"])
books
```
